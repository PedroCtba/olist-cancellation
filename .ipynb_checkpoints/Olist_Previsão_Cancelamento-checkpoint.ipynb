

# "Previso de Cancelamento de Pedido"

**O Problema:**

Algumas compras so canceladas pelos clientes oque diminui o faturamento da empresa e muitas vezes custeia  estrutura at o momento do estorno

**Como resolver isso com data science?**

Criar um algortimo que tenha capacidade analtica para elencar quais clientes tem "hiptese de cancelar", irei fazer isso usando atributos das ordens como "Target". Caso o algortimo seja preciso em dizer qual ordem foi cancelada com atributos que existem antes do cancelamento, ele pode gerar uma lista de "possveis canceladores".



*Obs: seguirei uma abordagem CRISP, ento vou resolver o mesmo problema do incio at o deploy em fases passando vrias vezes de maneira cclica em cada fase, at conseguir uma preciso que julgue satisfatria ou at "quase impossvel" de aumentar.*
